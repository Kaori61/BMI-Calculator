# BMI Calculator
BMI range referenced from https://www.calculator.net/bmi-calculator.html

```python
 #Height and wieght input form user
Height = int(input("Height in cm:   "))

Weight = int(input("Weight in kg:   "))
        
 # Calculating BMI    
BMI = Weight / (Height*0.01)**2
print(f"BMI: {BMI:.2f}")

 # Weight classification besed on BMI 
if BMI>0:
    if (BMI<18.5):
        print("You are under weight.")
    elif (BMI<24.9): 
        print("You are normal weight.")
    elif (BMI<29.9):
        print("Your are over weight.")
    elif (BMI<34.9):
        print("You are obese class 1.")
    elif (BMI<39.9):
        print("You are obese class 2.")
    elif (BMI<40):
        print("You are obese class 3.")
    else:
        print("Enter valid input")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
 #crating range for the graph
heights_cm = np.arange(140,201,1) #height in cm (140cm - 200cm, step size 1cm)
weights_kg = np.arange(20,150,1)

 #converting height to meters squared 
heights_squared = (heights_cm * 0.01)**2

 #creating combination of height and weight array
H, W =np.meshgrid(heights_squared, weights_kg)

BMI = W / H 
```

```python
 #plotting the map
    
plt.figure(figsize=(10,6))
contour = plt.contourf(heights_cm, weights_kg, BMI, levels =np.arange(10,41,2), cmap='coolwarm')
cbar =plt.colorbar(contour)
cbar.set_label('BMI', fontsize = 12)

 # Add labels and title

plt.title("BMI Chart", fontsize = 16)
plt.xlabel("Height (cm)", fontsize = 12)
plt.ylabel("Weight (kg)", fontsize = 12)
plt.grid(alpha = 0.3)
```
